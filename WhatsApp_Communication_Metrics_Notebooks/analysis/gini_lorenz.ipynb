Gini = 0 means Perfect equality (all contacts equally active) and < 0.5 means less inequality(Relatively balanced distribution) and >0.5 means high inequality, 
Gini = 1 means Perfect inequality (only one contact sends all messages)


The Lorenz curve complements the Gini index by plotting the cumulative percentage of messages against the cumulative percentage of contacts.

The diagonal line = perfect equality (everyone gets equal messages).


A bowed blue curve = more inequality.

The area between the two = inequality (accourding to gini value).
